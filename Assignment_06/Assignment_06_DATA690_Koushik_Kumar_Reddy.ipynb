## Let us got through each column to analyse the dataframe 

### 1. In the ***Census Year*** column, we can see that the 24th field has a asterisk printed along with the year. The data cannot be accessed in a smooth way if it exists, hence it should be cleansed. 

### 2. In the ***Census Cost*** column, the costs are mentioned in numbers as well as in string values. Again, whenever we would want to access the data, it will be an issue. So, it is better to change everything into integer values.

### 3. In the ***Average Cost Per Person*** column, there is no standard representation of the cost per person, hence only one way of representation, i.e either in dollars or in cents will be a better way of representation to access the data properly without any issues.

5. Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string. Test the function by using test strings (for example, "1989*", "20*10").

```
def cleanse_year(k):
  a = k.replace("*","")
  print(a)
  return a
cleanse_year("1989*")
```

6. Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.

```
df1["Census Year"] = df1["Census Year"].apply(cleanse_year)
df1
```

7. Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function).

```
df1["Total Population"] = df1["Total Population"].apply(cleanse_year)
df1
```

```
cleansed_cesus_cost=[]
def cleanse_census_cost(k):
  values = k.replace(",","").replace("$","")
  if  k.find("4.5") != -1:         
        values = "45" + "0" * 8
  elif k.find("Billion") != -1:    
        values = "13" + "0" * 9
  cleansed_cesus_cost.append(values)
  return values
df1["Census Cost"] = df1["Census Cost"].apply(cleanse_census_cost)
df1
```

```
def cleanse_avg_cost(k):
  if 'cents' in k:
    avg_cost = k.replace("cents","")
    avg_cost = float(avg_cost)/100
  else:
    avg_cost = k.replace("$","")
  return avg_cost
df1["Average Cost Per Person"] = df1["Average Cost Per Person"].apply(cleanse_avg_cost)
df1
```
