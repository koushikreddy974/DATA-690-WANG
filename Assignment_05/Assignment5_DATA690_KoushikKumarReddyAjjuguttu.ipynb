##Creation Of Smaller Dataframe with the given 9 Variables or Columns

```
df1 = df[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]
print(df1.head())
```

#Definitions of the 9 Variables/Columns

1. UNTID:- ***UNITID*** is a Unique Identification number    assigned to postsecondary institutions.
2. INSTNM:- ***INSTNM*** represents the name of the institution.
3. CITY:- ***CITY*** includes all the cities in which each Institution is located.
4. STABBR:- Represents the location of the institution.
5. ZIP:- Includes the area code or ***Zip*** code of that particular institution.
6. ADM_RATE:- ***ADM_RATE*** includes the admissions rate at each campus.
7. UGDS:- ***UGDS*** includes the total enrollment of Undergraduate Degree-Seeking students.
8. TUITIONFEE_IN:- Includes the in-state tutiton fee details of the students
9. MN_EARN_WNE_P*:- MN_EARN_WNE_P* earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year.

##Finding UMBC

```
df2 = pd.DataFrame(columns=['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10'])
```

```

for index in range(len(df1)):
  if df1.loc[index,'INSTNM'] == 'University of Maryland-Baltimore County':
    print(df1.loc[index])
    df2 = df2.append(df1.loc[index])
```

##Finding University of Maryland College Park



```
for index in range(len(df1)):
  if df1.loc[index,'INSTNM'] == 'University of Maryland-College Park':
    print(df1.loc[index])
    df2 = df2.append(df1.loc[index])
```

##Finding Loyola College Maryland

```
for index in range(len(df1)):
  if df1.loc[index,'INSTNM'] == 'Loyola University Maryland':
    print(df1.loc[index])
    df2 = df2.append(df1.loc[index])
```

##Finding Johns Hopskins

```
for index in range(len(df1)):
  if df1.loc[index,'INSTNM'] == 'Johns Hopkins University':
    print(df1.loc[index])
    df2 = df2.append(df1.loc[index])
```

#Calculating the min, max, and average in-state tution of the above four institutions.

```
print(df2['TUITIONFEE_IN'].min())
print(df2['TUITIONFEE_IN'].max())
print(df2['TUITIONFEE_IN'].mean())
```

# Colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia
